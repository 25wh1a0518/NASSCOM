```python
fruits = ['apple','banana','cherry','apple']
print("original list:",fruits)
print("first fruit:",fruits[0])
```

```python
fruits = ['apple','banana','cherry','apple']
fruits.append('orange')
print("after appending:",fruits) #adding an element
#removing a item
fruits.remove('apple')
print("after removing:",fruits)
#length of string
print("length of the string:",len(fruits))
```

```python
# tuple

dimension = (10,20,30)
print("Tuple:",dimension)
# accesing an element
print("third element",dimension[2])
```

```python
# sets

numbers = {1,4,3,5,1,3}
print("set(no duplicates):",numbers)
#adding a item
numbers.add(7)
print("after adding:",numbers)
#removing an item
numbers.remove(4)
print("after removing",numbers)

```

```python
#set operation

set1 = {1,2,3,4,5}
set2 = {2,3,8}
print("Union:",set1 | set2)
print("Intersection:",set1&set2)
print("Difference",set1-set2)
```

```python
# dictionaries
student = {
    'name': 'Alice',
    'age': 21,
    'course': 'Python'
}
print("Dictionary:", student)
#adding grade
student['grade'] = 'A'
print("after adding:",student)
#deleteing
del student['age']
print("After deleting age:", student)

# Loop
for key, value in student.items():
    print(f"{key}: {value}")
keys = ['id', 'name', 'marks']
values = [101, 'John', 88]
student_info = dict(zip(keys, values))
print("Student Info from zip:", student_info)

```
