```python

def greet():
    print("Hello! Welcome to Python Functions.")

# Calling the function
greet()
```

```python
#functions with parameter
def greet_user(name):
    print(f"Hello, {name}!")

greet_user("Siri")
greet_user("pranavi")
```

```python
#function with default parameter
def greet_user(name="Guest"):
    print(f"Welcome, {name}!")

greet_user()  #default
greet_user("Ravi")
```

```python
#USING ARUGUMENTS
def print_names(*names):
    print("Names passed:", names)
    for name in names:
        print("Hello", name)

print_names("siri", "virat", "Charvi")
```

```python
# keywordarguments
def print_info(**info):
    print("User Info:")
    for key, value in info.items():
        print(f"{key} = {value}")

print_info(name="Alice", age=25, city="Delhi")
```

```python
 #COMBINING *args AND **kwargs

def show_details(*args, **kwargs):
    print("Args:", args)
    print("Kwargs:", kwargs)

show_details(1, 2, 4, name="siri", city="hyderabad")
```

```python
#FUNCTION NESTING (CALL ONE FUNCTION INSIDE ANOTHER)

def multiply(a, b):
    return a * b

def square(n):
    return multiply(n, n)

print("Square of 4:", square(4))
```
