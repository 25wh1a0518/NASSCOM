```python
name = "Alice"
age = 25
height = 5.6

# Print values
print("Name:", name)
print("Age:", age)
print("Height:", height)

# Variables are dynamically typed
name = 123  # Now 'name' holds an integer
print("New value of name:", name)
```

```python
# Integer
a = 15
print(type(a))

# Float
b = 2.34
print(type(b))
# String
c = "Python"
print(type(c))

# Boolean
d = True
print(type(d))

# List
e = [1, 2, 3, "four"]
print(type(e), e)

# Dictionary
f = {"name": "Alice", "age": 25}
print(type(f), f)
```
